```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
```

```python

categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health']
```

```python
data = {'Date': pd.date_range('2024-09-01', periods=500), 'Category': [random.choice(categories) for _ in range(500)],
        'Likes': np.random.randint(0, 10000, 500)}
```

```python
print(pd.DataFrame(data))
```

```python
df = pd.DataFrame(data)
print(df.value_counts('Category'))
print(df.head())
df.describe()
df.info()
```

```python
clean_df = df.dropna()
clean_df.drop_duplicates(inplace=True)
clean_df['Date'] = pd.to_datetime(clean_df['Date'])
clean_df['Likes'] = pd.to_numeric(clean_df['Likes'])
print(clean_df)
```

```python
sns.histplot(df['Likes'], bins=30, kde=True)
plt.show()
sns.boxplot(data, x='Category', y='Likes')
plt.show()
```

```python
print('The overall mean is:', df['Likes'].mean())
category_mean_likes = df.groupby('Category')['Likes'].mean()
print("Mean of Likes by Category:\n", category_mean_likes)
```
